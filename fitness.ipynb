```python
import numpy as np
import plotly.graph_objects as go
```

```python
def cell_fitness(type,s,k):
    if type == 'log': return 1 + s * np.log(1+k)
    elif type == 'linear': return 1 + s * k
    elif type == 'power': return (1+s)**k
```

```python
def plot_fitness_functions(functions, n=200):
    x = np.linspace(0, n, n+1)
    
    fig = go.Figure()

    for function_type, s in functions:
        y = cell_fitness(function_type, s, x)
        fig.add_trace(go.Scatter(x=x, y=y, mode='lines', name=f'{function_type}, s={s}',
                                 hovertemplate='ecDNA copy number: %{x}<br>Fitness: %{y}'))

    fig.update_layout(title='Fitness functions',
                      xaxis_title='ecDNA copy number',
                      yaxis_title='Fitness',
                      showlegend=True,
                      plot_bgcolor='rgba(0,0,0,0)')

    fig.show()

# Example usage:
functions = [('log', 0.2), ('log',0.1), ('linear',0.01)]
plot_fitness_functions(functions)
```
