```python
# Création du dictionnaire cell_counts_p4 avec des clés entières
cell_counts_p4 = df_p4['ecCounts'].round().apply(int).value_counts().to_dict()
cell_counts_p15 = df_p15['ecCounts'].round().apply(int).value_counts().to_dict()
```

```python
file_path_p4 = 'data/cell_counts_p4.yaml'
file_path_p15 = 'data/cell_counts_p15.yaml'
save_yaml(cell_counts_p4, file_path_p4)
save_yaml(cell_counts_p15, file_path_p15)
cell_counts_p4 = load_yaml(file_path_p4)
cell_counts_p15 = load_yaml(file_path_p15)
```

```python
plot_histogram_dict(cell_counts_p4, plot_color='turquoise', bin_size=5, title='P4 cell counts')
```

```python
plot_histogram_dict(cell_counts_p15, bin_size=10, title='P15 cell counts', plot_color='deepskyblue')
```

```python
import numpy as np
print('Mean ecDNA count at passage 4:', np.mean(df_p4['ecCounts']))
print('Mean ecDNA count at passage 15:', np.mean(df_p15['ecCounts']),f'\n')
print('Variance of ecDNA count at passage 4:', np.var(df_p4['ecCounts']))
print('Variance of ecDNA count at passage 15:', np.var(df_p15['ecCounts']))
```
