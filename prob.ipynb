```python
import numpy as np
import plotly.graph_objs as go

def probability_death(s, N=1000, fitness='log'):
    if fitness == 'log':
        f = 1 + s * np.log(1 + 1)
    elif fitness == 'linear':
        f = 1 + s * 1
    else:
        raise ValueError('fitness not defined')

    p = f / (f + (N - 1) * 1)  # suppose that f0 = 1
    return 1 / (1 + (N - 1) * p)

def plot(fitness_functions=['log','linear']):
    # Define s values from 0 to 2
    s_values = np.linspace(0, 2, 100)
    fig = go.Figure()

    for fitness in fitness_functions:
        probabilities = [probability_death(s=s, fitness=fitness) for s in s_values]

        scatter_plot = go.Scatter(x=s_values, 
                                  y=probabilities, 
                                  mode='markers', 
                                  marker=dict(size=8), 
                                  name=f'{fitness} fitness')
        
        fig.add_trace(scatter_plot)

    fig.layout = go.Layout(title=f'Probability that the first single cell with ecDNA dies vs. s',
                    xaxis=dict(title='s'),
                    yaxis=dict(title='Probability'))


    fig.show()
plot()
```
